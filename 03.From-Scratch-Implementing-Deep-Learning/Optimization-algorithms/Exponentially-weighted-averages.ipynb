# 指数加权平均数

```python
import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider  

x = np.arange(0, 100, 0.3)

y = 10 -(x/10 - 5) ** 2 + np.random.randn(x.shape[0]) * 5

ts_data = pd.Series(data=y, index=x)
```

```python
ts_data.plot(figsize=(15, 3))
plt.show()
```

```python
def ewa(belta):
    v_ = 0
    belta = belta
    new_data = []
    for v in ts_data:
        new_data.append(v_ * belta + v * (1 - belta))
    ts_new_data = pd.Series(data=new_data, index=x)
    ts_new_data.plot(figsize=(15, 3))
    plt.plot(ts_new_data/(1 - belta**(x+1)))
    ts_data.plot()
    plt.show()
```

```python
ewa(0.9)
ewa(0.8)
ewa(0.7)
ewa(0.5)
ewa(0.4)
ewa(0.3)
ewa(0.1)
```
