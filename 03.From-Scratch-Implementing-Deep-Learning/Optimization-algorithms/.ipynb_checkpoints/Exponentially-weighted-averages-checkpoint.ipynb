# 指数加权平均数

```python
import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider  

x = np.arange(0, 100, 0.3)

y = 10 -(x/10 - 5) ** 2 + np.random.randn(x.shape[0]) * 5

ts_data = pd.Series(data=y, index=x)
```

```python
ts_data.plot(figsize=(15, 3))
plt.show()
```

```python
def ewa(belta):  
    v_ = 0  
    new_data = []  
    for v in ts_data:  
        v_ = v_ * belta + v * (1 - belta)  
        new_data.append(v_)  
    ts_new_data = pd.Series(data=new_data, index=ts_data.index)  
    
    ax.clear()  
    ts_data.plot(ax=ax, label='Original Data')  
    ts_new_data.plot(ax=ax, label='EWA', color='orange')  
    ax.axhline(0, color='black', lw=0.8, ls='--')  
    ax.set_title(f'Exponential Weighted Average with belta = {belta:.2f}')  
    ax.legend()  
    plt.draw()  
```

```python
# 设置图形和滑块  
fig, ax = plt.subplots(figsize=(15, 5))  
plt.subplots_adjust(bottom=0.25)  

# 默认 belta 值  
initial_belta = 0.5  
ewa(initial_belta)  

# 添加滑块  
ax_belta = plt.axes([0.2, 0.1, 0.65, 0.03])  
slider_belta = Slider(ax=ax_belta, label='Belta', valmin=0.01, valmax=1, valinit=initial_belta, valstep=0.01)  

# 当滑块值改变时更新图形  
slider_belta.on_changed(ewa)  

plt.show()  
```
